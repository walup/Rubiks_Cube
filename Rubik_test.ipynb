```python
nTurns = 1
axis = Axis.Y
counterClockwise = False
cube.rotateWholeCube(nTurns, counterClockwise, axis)

cube.drawCube(-50,30,0)
```

# Solving example 

I won't teach how to solve the thing at great detail, at this point i don't even remember the algorithms very well, i just know how to do it by muscular memory, so i will have to reverse engineer them. 

Notation for the moves is quite simple F means you rotate the front face 1 time clockwise. If you write F', it is the same thing but you are going to rotate counterclockwise. Rotations for other faces use the following letters:

B: back face
R: right face
L: left face
U: up face
D: down face

This is all standard. An additional rule is that if you want to rotate say the front face 2 times clockwise, you can write that as F2. 

Ok let's first scramble the cube, i will do that using the moves:

F, B, R2, L2, F, R', L', D2, U, B2, R, L, F, B

which i just randomly wrote. 

```python
cube = RubiksCube()
cube.makeMove("F")
cube.makeMove("B")
cube.makeMove("R2")
cube.makeMove("L2")
cube.makeMove("F")
cube.makeMove("R'")
cube.makeMove("L'")
cube.makeMove("D2")
cube.makeMove("U")
cube.makeMove("B2")
cube.makeMove("R")
cube.makeMove("L")
cube.makeMove("F")
cube.makeMove("B")


cube.drawCube(-50,30,0)
```

And there it is, we have the first two layers completed. The next step is to get a dog-shape on the top layer, and for this we have to apply some algorithms. 

The first is the one that sounds FRURUF

```python
cube.makeMove("F")
cube.makeMove("R")
cube.makeMove("U")
cube.makeMove("R'")
cube.makeMove("U'")
cube.makeMove("F'")

cube.drawCube(-50,30,0)
```

Now this is not yet a dog, let's see our yellow layer:

```python
cube.drawCube(-80,90,0)
```

There's actually a specific algorithm that you can use for this figure and complete the layer, but i don't remember it. I used to know it a long time ago, but it has faded away. So, what i always do is apply the same algorithm until i get a dog figure.

The next algorithm sounds something like RURURUR



```python
#We will repeat this until we get a dog. 
cube.makeMove("R")
cube.makeMove("U")
cube.makeMove("R'")
cube.makeMove("U")
cube.makeMove("R")
cube.makeMove("U2")
cube.makeMove("R'")
```

```python
cube.drawCube(-80,90,0)
```

```python
#Apply it again
cube.makeMove("R")
cube.makeMove("U")
cube.makeMove("R'")
cube.makeMove("U")
cube.makeMove("R")
cube.makeMove("U2")
cube.makeMove("R'")

cube.drawCube(-80,90,0)
```

Ok, that's our dog. It is even in the right position, which means we were quite lucky. Now you just have to apply the same algorithm again and magic will happen. 

```python
cube.makeMove("R")
cube.makeMove("U")
cube.makeMove("R'")
cube.makeMove("U")
cube.makeMove("R")
cube.makeMove("U2")
cube.makeMove("R'")

cube.drawCube(-50,30,0)
```

Great, we are almost done. We are so lucky that we are even going to be able to skip one step. 

Let's align the last layer and complete the orange face. This orange face has to be in the back for the final steps so we will rotate the whole cube.

```python
cube.makeMove("U'")
cube.drawCube(-50,30,0)
```

```python
cube.rotateWholeCube(1, True, Axis.Z)
cube.drawCube(-50,30,0)
```

Now we apply the final algorithm as many times as necessary:

```python
cube.makeMove("R")
cube.makeMove("U'")
cube.makeMove("R")
cube.makeMove("U")
cube.makeMove("R")
cube.makeMove("U")
cube.makeMove("R")
cube.makeMove("U'")
cube.makeMove("R'")
cube.makeMove("U'")
cube.makeMove("R2")
```
